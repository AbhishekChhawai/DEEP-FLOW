```python
plt.figure(figsize=(6,3))
plt.scatter(df['votes'], df['Rating'], alpha=0.5, color='r')
plt.title('Votes vs Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.scatter(df['approx_cost(for two people)'], df['Rating'], alpha=0.5, color='green')
plt.title('Cost for Two vs Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```
